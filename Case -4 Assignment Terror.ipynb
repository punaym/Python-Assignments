```python
import pandas as pd
import numpy as np
```

```python
terror = pd.read_csv("terror.csv",encoding="Latin")

print(terror.head())
print(terror.columns)
print(terror.describe())
print(terror.shape)
```

```python
#Q.1) How many attacks happened in India?
len(terror.query("country_txt == 'India'"))
```

```python
# Q.2) How many attacks happened in India and upto 3 people were killed? 
len(terror.query("country_txt == 'India' and nkill <= 3"))
```

```python
#Q.3) Extract the city and summary for attacks above

terror.query("country_txt == 'India' and nkill <= 3")[['country_txt','city','summary']]
```

```python
#Q.4) In a single terror incident in India, find out top 5 cities by number killed

terror[terror.country_txt=='India'][['city','nkill','nwound']].sort_values(by = "nkill",ascending = False).head()
```

```python
#Q.5) In a single terror incident in India, find out top 5 cities by number killed and wounded 
terror[terror.country_txt=='India'][['city','nwound','nkill']].nlargest(5, 'nwound')
```

```python
#Q.6) How many attacks were successful that were suicide attacks?
len(terror.query("success == 1 and suicide == 1"))
```

```python
#Q.7) label all the incidents where the number killed was more than 5 as severe. 
terror['label_kill']=['severe' if i > 5.0 else 'not_severe' for i in terror.nkill]
terror['label_kill']
```

```python

def func(x):
           x['Label_Suicide'] = ["Sucide_Success" if( x.iloc[i].suicide == 1 and x.iloc[i].success == 1)   else "Suicide_NotSuccess" for i in list(range(0,len(x)))]
           return x['Label_Suicide']
          
func(terror)

```

```python
'''
Q.9) Create a new category representing if the incident occured in Afghanistan, Pakistan or India 
as one level of the category and all the other countries as another level
'''
terror['category']=['Asia' if (terror.iloc[i].country_txt=='Afghanistan' 
                               or terror.iloc[i].country_txt=='Pakistan' 
                               or terror.iloc[i].country_txt=='India') else 'ROW' for i in list(range(0,len(terror)))]
print(terror['category'].head())
```

```python
#Q.10) How many incidents happened in Af-Pak-India vs ROW?
print("No. of Incidents in Afghanisthan, Pakistan and India is: ", len(terror.query("category=='Asia'")))
len(terror.query("category=='ROW'"))
round(len(terror.query("category=='Asia'"))*100/len(terror.query("category =='ROW'")),2)
```

```python
'''
Q.11)	List the number of suicides attacks and average kills by Af-Pak-India vs ROW. Rename columns 
in the output as Average_Kills and Number_Incidents.
'''

print(terror.query("suicide == 1").groupby('category')[['nkill']].agg(['count','mean']).rename(columns = {'count' : "Number_Incidents",'mean':"Average_kills"}))
```
