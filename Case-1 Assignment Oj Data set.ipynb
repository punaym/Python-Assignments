```python
#16 Find:
#a.	Mean and std deviation of the income
#b.	For income greater than or equal to 10.5, find the mean income
#c.	For each brand having price >=2.5 find the mean, median, sd of the log of income

print(round(np.mean(oj.INCOME),2))
print(round(np.std(oj.INCOME),2))

mi= oj.query("INCOME >= 10.5")[["INCOME"]]
print(np.mean(mi))

i= oj.query("price >= 2.5")[["INCOME"]]
print(np.mean(np.log(i)))
print(np.median(np.log(i)))
print(np.std(np.log(i)))
```
