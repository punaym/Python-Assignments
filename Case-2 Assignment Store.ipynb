```python
import pandas as pd
import numpy as np
```

```python
store = pd.read_csv("Store.csv",encoding="LATIN")
store
```

```python
# How many unique cities are the orders being delivered to

store.City.unique()
```

```python
# What is the total quantity sold in the East Region?

tot_quantity= store.query('Region == "East"')[['Quantity']]
print("total", np.sum(tot_quantity))
```

```python
# Find the sum of the quantity sold in the East Region

tot_quantity= store.query('Region == "East"')[['Quantity']]
print("Sum of", np.sum(tot_quantity))
```

```python
# In the south region sort the sales in decreasing order

store.query('Region == "South"')[["Sales"]].sort_values(by='Sales',ascending= False)
```

```python
# Find the mean of quantity for every region
store.groupby('Region').Quantity.mean()
```

```python
# Find the mean of sales for every category

store.groupby('Category').Sales.mean()
```

```python
# Find the max, min, sum of sales and profit for every category
store.groupby('Category').agg({'Sales':['max','min','sum'], 'Profit':['max','min','sum']})
```

```python
# Find sum of sales and max profit for every segment

store.groupby('Segment').agg({'Sales':['sum','max']})
```

```python
# For every segment find the mean of the discount

store.groupby('Segment').agg({'Discount':['mean']})
```

```python
# For every segment find the most profitable customers
def get_cust(profit):
    profit = profit.max()
    return store[store.Profit == profit]['Customer Name']
store.groupby('Segment').agg({'Profit':[get_cust, 'max']})
```

```python
# What are the top 5 categories that give maximum profit?

store.groupby('Category').agg({'Profit':['max']})
```

```python
# What is the Total Sales, Quantity, Discount, Profit across Total US.
store.groupby('Country')[['Sales','Quantity','Discount','Profit']].agg(['sum'])
    
```

```python
# How many times has it taken more than 5 days from placing an order to shipping
days = pd.to_datetime(store['Ship Date']) - pd.to_datetime(store['Order Date'])
print("Number of Times has it taken more than 5 days:  ", sum(days > '5 days'))
```

```python
# Find the total number of orders in every category which has been shipped with a duration > 5 days

store['days'] = pd.to_datetime(store['Ship Date']) - pd.to_datetime(store['Order Date'])

def sum1(a):
    return sum(a <"5 days")

            
store.groupby('Category').agg({'days':[sum1]})
```

```python
# What’s the percentage of items which has been shipped within 5 days
store['days'] = pd.to_datetime(store['Ship Date']) - pd.to_datetime(store['Order Date'])

def sum1(a):
    return (sum(a <"5 days")/sum(days <"5 days")*100)

            
store.groupby('Category').agg({'days':[sum1]})
```

```python
# What’s the percentage of items which has been shipped after 5 days

store['days'] = pd.to_datetime(store['Ship Date']) - pd.to_datetime(store['Order Date'])

def sum1(a):
    return (sum(a >"5 days")/sum(days> "5 days")*100)

            
store.groupby('Category').agg({'days':[sum1]})
```
